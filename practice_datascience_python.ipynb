```python
import numpy as np
import matplotlib.pyplot as plt
rolls=[]
for k in range(100000):
    rolls.append(random.choice([1,2,3,4,5,6]))

plt.hist(rolls, bins=np.linspace(0.5,6.5,7))

# increase range from 100 to 100000 smothnes out the distribution due to law of large numbers
```

```python
# rolling 10 independent die
```

```python
ys=[]
for rep in range(10000):
    y=0
    for k in range(10):
        x=random.choice([1,2,3,4,5,6])
        y+=x  # y is a new random variable which is the sum of rolls from 10 independent dies
    ys.append(y)

print(ys, min(ys),max(ys))
```

```python
plt.hist(ys)
# change range from 100 to 10000 distribution normal (central limit theorem)
```

```python
# Simulating 2-D random walk
n=100   # increase steps of random walk  from 5 to 100
X_0=np.array([[0],[0]])
delta_X=np.random.normal(0,1,(2,n))
X=np.concatenate((X_0,np.cumsum(delta_X, axis=1)),axis=1) # random walk start at 0,0
plt.plot(X[0],X[1], 'ro-')
```

```python
# Brief introduction to matplotlib

x=np.linspace(0,10,20)
y1=x**2
y2=x**3
plt.plot(x,y1,'bo-', linewidth=2, markersize=12, label="First")
plt.plot(x,y2,'gs-', linewidth=2, markersize=12 , label = "Second")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(loc="upper left")
xmin=-0.5 
xmax=10
ymin=-5 
ymax=105
plt.axis([xmin,xmax,ymin,ymax])
plt.savefig("myplot.pdf")
```
