```python
# Simulating 2-D random walk
n=100   # increase steps of random walk  from 5 to 100
X_0=np.array([[0],[0]])
delta_X=np.random.normal(0,1,(2,n))
X=np.concatenate((X_0,np.cumsum(delta_X, axis=1)),axis=1) # random walk start at 0,0
plt.plot(X[0],X[1], 'ro-')
```

```python
# Linear regression 
# Generate data

from scipy import stats

n=100
beta_0 = 5
beta_1 = 2
np.random.seed(1)

x= 10*stats.uniform.rvs(size=n)
y= beta_0 + beta_1*x + stats.norm.rvs(loc=0, scale = 1, size=n)  # loc is mean, scale is std 


plt.figure()
xx = np.array([0 , 10])
plt.plot(x,y,"bo", ms=5)
plt.plot(xx,beta_0 + beta_1*xx)
plt.xlabel("x")
plt.ylabel("y")

```

```python
# least square estimation
rss = []
slopes=np.arange(-10,15,0.01)
for slope in slopes:
    rss.append(np.sum((y-beta_0-slope*x)**2))

ind_min = np.argmin(rss)

print(f" Estimate for slope is {slopes[ind_min]}")
plt.figure()
plt.plot(slopes, rss)
plt.xlabel("slopes")
plt.ylabel("rss")
```

```python
import statsmodels.api as sm 
X=sm.add_constant(x)
model=sm.OLS(y,X)
est=model.fit()
print(est.summary())
```

```python
# Classification
# Generate data for classification from two different normal distribution centered differently on x axis

def gen_data(h,sd1,sd2,n):
    x1=stats.norm.rvs(-h,sd1,n)
    y1=stats.norm.rvs(0,sd1,n)
    x2=stats.norm.rvs(h,sd2,n)
    y2=stats.norm.rvs(0,sd2,n)
    return(x1,y1,x2,y2)

(x1,y1,x2,y2)= gen_data(1,1,1.5,1000)

def plot_data(x1,y1,x2,y2):
    plt.figure()
    plt.plot(x1,y1,"o", ms=2)
    plt.plot(x2,y2,"o", ms=2)
    plt.xlabel("$X_1$")
    plt.ylabel("$X_2$")

plot_data(x1,y1,x2,y2)
```

```python
# Logistic regreession

from sklearn.linear_model import LogisticRegression
clf=LogisticRegression()

X=np.vstack((np.vstack((x1,y1)).transpose(), np.vstack((x2,y2)).transpose()))
print(X.shape)
n=1000
y=np.hstack((np.repeat(1,n),np.repeat(2,n)))
print(y.shape)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, train_size=0.5, random_state =1)
X_train.shape
X_test.shape

clf.fit(X_train,y_train)
clf.score(X_test,y_test)

clf.predict_proba(np.array([-2,0]).reshape(1, -1)) # you can try predicting without reshape and it will tell you how to reshape
```

```python
# Compute and plot predictibe probabilities for grid and not single (x1,x2)

def plot_probs(ax, clf, class_no):
    xx1, xx2 = np.meshgrid(np.arange(-5, 5, 0.1), np.arange(-5, 5, 0.1))
    probs = clf.predict_proba(np.stack((xx1.ravel(), xx2.ravel()), axis=1))
    Z = probs[:,class_no]
    Z = Z.reshape(xx1.shape)
    CS = ax.contourf(xx1, xx2, Z)
    cbar = plt.colorbar(CS)
    plt.xlabel("$X_1$")
    plt.ylabel("$X_2$")

# plotting code
plt.figure(figsize=(5,8))
ax = plt.subplot(211)
plot_probs(ax, clf, 0)
plt.title("Pred. prob for class 1")
ax = plt.subplot(212)
plot_probs(ax, clf, 1)
plt.title("Pred. prob for class 2");
```
